```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 

rg = RandomForestRegressor(n_jobs=-1,n_estimators=2)
preds = []
rmsl = []
#for i in range(1,df2.shape[0]-1):
start = df2.shape[0]-100
end = df2.shape[0]-2
for i in range(start,end):
    train = df2.iloc[0:i,:]
    tr_labels = labels[0:i]
    test = df2.iloc[i+1,:]
    rg.fit(train,tr_labels)
    preds.append(rg.predict(test)[0])
    actuals = labels[start+1:i+2]
    rms = np.sqrt(mean_squared_error(actuals, preds))
    rmsl.append(rms)
    if i%100 == 0:
        print i
```
